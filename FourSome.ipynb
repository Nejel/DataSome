```python
#feat_TerminationFact = tf.feature_column.numeric_column('TerminationFact')
#feat_Terminationdate = tf.feature_column.numeric_column('Terminationdate')
#feat_TerminationReason = tf.feature_column.categorical_column_with_hash_bucket('TerminationReason',hash_bucket_size=30)
#feat_TermReasonCategory = tf.feature_column.categorical_column_with_hash_bucket('Termreasoncategory',hash_bucket_size=30)

feat_Y = tf.feature_column.numeric_column('Y')
feat_M = tf.feature_column.numeric_column('M')
feat_EmplId = tf.feature_column.numeric_column('EmplId')
feat_FullNameEng = tf.feature_column.categorical_column_with_hash_bucket('FullNameEng',hash_bucket_size=2300)
feat_Status = tf.feature_column.categorical_column_with_hash_bucket('Status',hash_bucket_size=3)
feat_VacationDaysCollected = tf.feature_column.numeric_column('VacationDaysCollected')
feat_TopDept = tf.feature_column.categorical_column_with_hash_bucket('TopDept',hash_bucket_size=30)
feat_DepFull = tf.feature_column.categorical_column_with_hash_bucket('DepFull',hash_bucket_size=470)
feat_StaffCountry = tf.feature_column.categorical_column_with_hash_bucket('StaffCountry',hash_bucket_size=5)
feat_Location = tf.feature_column.categorical_column_with_hash_bucket('Location',hash_bucket_size=35)
feat_HomeOffice = tf.feature_column.numeric_column('HomeOffice')
feat_StaffTypeName = tf.feature_column.categorical_column_with_hash_bucket('StaffTypeName',hash_bucket_size=3)
feat_HC = tf.feature_column.numeric_column('HC')
feat_TopDptReorg = tf.feature_column.categorical_column_with_hash_bucket('Topdptreorg',hash_bucket_size=35)
feat_Division = tf.feature_column.categorical_column_with_hash_bucket('Division',hash_bucket_size=3)
feat_Region = tf.feature_column.categorical_column_with_hash_bucket('Region',hash_bucket_size=3)
feat_Hiredate = tf.feature_column.numeric_column('Hiredate')
feat_LoS = tf.feature_column.numeric_column('LoS')
feat_DoB = tf.feature_column.numeric_column('DoB')
feat_Age = tf.feature_column.numeric_column('Age')
feat_Gender = tf.feature_column.categorical_column_with_hash_bucket('Gender',hash_bucket_size=3)
feat_Grade = tf.feature_column.numeric_column('Grade')
feat_Manager = tf.feature_column.numeric_column('Manager')
feat_SPAN = tf.feature_column.numeric_column('SPAN')
feat_Performance = tf.feature_column.numeric_column('Performance')
feat_Talent = tf.feature_column.numeric_column('Talent')
feat_CR = tf.feature_column.numeric_column('CR')
feat_Averageage = tf.feature_column.numeric_column('Averageageintheteam')
feat_Agediffwithteam = tf.feature_column.numeric_column('Agediffwithteam')
feat_numberofteammates = tf.feature_column.numeric_column('numberofteammates')
feat_Managerage = tf.feature_column.numeric_column('Managersage')
feat_AgediffwithManager = tf.feature_column.numeric_column('AgediffwithManager')
feat_Differentlocationwithmanager = tf.feature_column.numeric_column('Differentlocationwithmanager')
feat_Overtimehourscumul = tf.feature_column.numeric_column('Overtimehourscumul')
feat_Overtimehourspermonth = tf.feature_column.numeric_column('Overtimehourspermonth')


feat_cols = [feat_Y,feat_M,feat_EmplId,feat_FullNameEng,feat_Status,feat_VacationDaysCollected,feat_TopDept,feat_DepFull,feat_StaffCountry,feat_Location,feat_HomeOffice,feat_StaffTypeName,feat_HC,feat_TopDptReorg,feat_Division,feat_Region,feat_Hiredate,feat_LoS,feat_DoB,feat_Age,feat_Gender,feat_Grade,feat_Manager,feat_SPAN,feat_Performance,feat_Talent,feat_CR,feat_Averageage,feat_Agediffwithteam,feat_numberofteammates,feat_Managerage,feat_AgediffwithManager,feat_Differentlocationwithmanager,feat_Overtimehourscumul,feat_Overtimehourspermonth]
#feat_cols = [feat_Y,feat_M,feat_EmplId,feat_FullNameEng,feat_Status,feat_VacationDaysCollected,feat_Terminationdate,feat_TerminationReason,feat_TermReasonCategory,feat_TopDept,feat_DepFull,feat_StaffCountry,feat_Location,feat_HomeOffice,feat_StaffTypeName,feat_HC,feat_TopDptReorg,feat_Division,feat_Region,feat_Hiredate]
```

```python
x_data = data.drop('TerminationFact',axis=1)
labels = data['TerminationFact']

x_data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,labels, test_size=0.33, random_state=101)


input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=100,num_epochs=1000,shuffle=True)
eval_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)

model = tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=1000)
```

```python
results = model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
predictions = model.predict(pred_input_func)
```
